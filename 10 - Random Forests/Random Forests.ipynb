### Random Forests

*** 

### Introduction:

### Random Forests are an advanced ensemble learning method that enhance the predictive performance and robustness of decision trees. While a decision tree is a single model prone to overfitting and high variance, a Random Forest constructs multiple decision trees using random subsets of the data and features, averaging their outputs to improve accuracy and control overfitting. 

### This method reduces variance without significantly increasing bias, leading to more reliable predictions. Although Random Forests offer superior predictive power and stability compared to individual decision trees, they are less interpretable due to their aggregated decision-making process.


### In this section, we will build a Random Forest model and utilize it on the Iris dataset.

***

### Importing libraries and data

```python
# Importing libraries
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing data
df = pd.read_csv('iris.csv')
df.head()
```

### Visualizing the Sepal Length and Sepal Width

```python
fig = px.scatter(df, x="SepalLengthCm", y="SepalWidthCm", color="Species", title="Sepal Length and Sepal Width of Irises")
fig.show("png")
```

### Visualizing the Petal Length and Petal Width

```python
fig = px.scatter(df, x="PetalLengthCm", y="PetalWidthCm", color="Species", title="Petal Length and Petal Width of Irises")
fig.show("png")
```

### Encoding the species column and building a Random Forest model

```python
# Features: Sepal length, Sepal width, Petal length, Petal width
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values

# Target variable: Species (currently in string form)
y = df['Species'].values

# Encode the target variable (species) into numeric classes
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Initialize the Random Forest model
random_forest = RandomForestClassifier()

# Set up K-Fold Cross Validation
kf = KFold(n_splits=3, shuffle=True)

# Perform Cross Validation
cv_scores = cross_val_score(random_forest, X, y_encoded, cv=kf)

# Print the cross-validation scores
print("Cross-validation scores: ", cv_scores)
print("Mean cross-validation score: ", cv_scores.mean())
```

*** 

### Conclusion:

### In this notebook, we applied a Random Forest model to the Iris dataset to classify the species of iris flowers based on four features: sepal length, sepal width, petal length, and petal width. 

### The mean cross-validation score was 0.94, indicating that the Random Forest model performs well in classifying the species of the iris flowers based on the given features.

### Overall, the Random Forest model demonstrated strong and consistent performance across different folds, showcasing its effectiveness and robustness in handling this classification task.
