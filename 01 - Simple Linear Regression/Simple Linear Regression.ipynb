### Simple Linear Regression

***

### Introduction: 

### Linear Regression is a fundamental statistical method used in machine learning for modeling the relationship between a dependent variable and one or more independent variables. 

### When dealing with a single independent variable, it is referred to as Simple Linear Regression. This technique aims to find the best-fit line that predicts the dependent variable based on the independent variable by minimizing the sum of the squared differences between the observed values and the values predicted by the linear model.

### The simplicity and interpretability of Linear Regression make it a popular choice for initial data analysis and predictive modeling, providing a baseline for more complex models and offering insights into the strength and nature of relationships within the data.

### This section focuses on training a Linear Regression model using the size of apartments in order to predict the price of apartments.

***

### Importing libraries and data

```python
# Importing libraries
import pandas as pd
import plotly.express as px
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing data
df = pd.read_excel('apartments.xlsx')
df.head()
```

### Visualizing the size and price of apartments

```python
fig = px.scatter(df, x="Size", y="Price", title="Size in m² and price of apartments")
fig.show("png")
```

### Building a Simple Linear Regression model using the size as a feature

```python
# Define features and target variable
X = df[['Size']]
y = df[['Price']]

# Instantiate model
model = LinearRegression()

# Train model
model.fit(X, y)

# Output intercept and coefficients
print('\nCoef: ' + str(model.coef_))
print('\nIntercept: ' + str(model.intercept_))
print('\nLinear Regression formula = Price =', str(model.coef_), '* Size in m² +', str(model.intercept_))

# Making a prediction for a 40m² apartment
y_pred = model.predict([[40]])
print('\nA 40m² apartment is estimated to cost:', y_pred)
```

### Visualizing the model

```python
fig = px.scatter(df, x="Size", y="Price", title="Size in m² and price of apartments", trendline = "ols")
fig.show("png")
```
