## Train/Test Split: Splitting input data and optimizing model training

## Context:

## Using a train/test split in machine learning is crucial for evaluating a model's performance on new, unseen data. By dividing the dataset into a training set and a test set, we can train the model on one portion (the training set) to help it learn patterns and relationships within the data. 

## Then, we test the model on the remaining portion (the test set) to see how well it predicts outcomes on data it hasn't encountered before. 

## This process ensures that the model can generalize it's learning to new data rather than just memorizing the training data, leading to more reliable and robust machine learning models that perform well in real-world scenarios.

## Importing libraries and data

```python
# Importing libraries
import pandas as pd
import plotly.express as px
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing data
df = pd.read_excel('apartments.xlsx')
df.head()
```

```python
# Missing values?
df.info()
```

```python
# Statistical information
df.describe()
```

## Visualizing the size and price of apartments

```python
fig = px.scatter(df, x="Size", y="Price", title="Size in m² and price of apartments")
fig.show("png")
```

## Train/Test Split using 75% of data for training and 25% of data for testing

```python
# Define features and target variable
X = df[['Size']]
y = df[['Price']]

# Performing the train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

# Joining the data for visualization
train_df = X_train.join(y_train, how = 'inner')
test_df = X_test.join(y_test, how = 'inner')

# Adding more information to the dataframe
train_df['Type'] = 'Training Data'
test_df['Type'] = 'Testing Data'
```

```python
# Visualizing the train test split using a scatter plot
split_viz = pd.concat([train_df, test_df], axis=0)

fig = px.scatter(split_viz, x='Size', y='Price', color = 'Type', title='Train Test Split Visualization')
fig.show("png")
```

## Training a Linear Regression model using the training data
## Making predictions using the testing data

```python
# Instantiate model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)

# Generating predictions and plotting the predictions against the actual results
y_pred = model.predict(X_test)

pred_viz = pd.DataFrame({'X_test': X_test.squeeze(), 'y_test': y_test.squeeze(), 'y_pred': y_pred.squeeze()})

fig = px.scatter(pred_viz, x='X_test', y='y_test', title='Actual vs. Predicted Data')
fig.add_scatter(x=pred_viz['X_test'], y=pred_viz['y_pred'], mode='lines', name='Predicted Line')
fig.show("png")
```

## Conclusion: 

## In this example, a Linear Regression model was trained with 75% of the original data. 

## Testing the model on the remaining 25% of the data revealed that the model was able to successfully generate a Linear Regression line on data it didn't see before.

## This indicates that the model can generalize it's learning to new data rather than just memorizing the training data
