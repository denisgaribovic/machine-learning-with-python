## Linear Regression with one variable - Predicting apartment prices using a simple dataset

## Importing libraries and the dataset

```python
# Importing libraries
from IPython.display import display, HTML
import warnings

import pandas as pd
import plotly.express as px
from sklearn.linear_model import LinearRegression

display(HTML('<style>.container { width:90% !important; }</style>'))
warnings.filterwarnings("ignore")
```

```python
# Importing data
df = pd.read_excel('apartments.xlsx')
df.head()
```

```python
# Missing values?
df.info()
```

```python
# Statistical information
df.describe()
```

## Visualizing data

```python
fig = px.scatter(df, x="Size", y="Price", title="Size in m² and price of apartments")
fig.show("png")
```

## Training a Linear Regression model using the size of apartments and making a prediction for the price

```python
# Instantiate model
model = LinearRegression()

# Define features and target variable
X = df[['Size']]
y = df[['Price']]

# Train model
model.fit(X, y)

# Output intercept and coefficients
print('Coef: ' + str(model.coef_))
print('Intercept: ' + str(model.intercept_))
print('Linear Regression formula = Price =', str(model.coef_), '* Size in m² +', str(model.intercept_))

# Making a prediction for a 40m² apartment
y_pred = model.predict([[40]])
print('A 40m² apartment is estimated to cost:', y_pred)
```

## Visualizing a Linear Regression line

```python
fig = px.scatter(df, x="Size", y="Price", trendline="ols", title="Size in m² and price of apartments (with Linear Regression line)")
fig.show("png")
```
